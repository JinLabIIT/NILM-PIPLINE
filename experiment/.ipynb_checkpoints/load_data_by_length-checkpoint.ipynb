```python
import pandas as pd
import json
import numpy as np
import math
```

```python
appliance = 'kettle'
length = 100
```

```python
df = pd.read_csv('..//data/dataset/dataset_x5_beta.csv')
```

```python
df
```

```python
l = np.array()[ (json.loads(app_power), contain_activation) for index, (app_power, contain_activation) in df[(df['appliance']==appliance) & (df['is_synthetic']==0) ][['app_power', 'contain_activation']].iterrows()]
```

```python
label = np.array([ t[1] for t in l ])
left_padding_size = math.floor((length-app_power.shape[1])/2)
right_padding_size = math.ceil((length-app_power.shape[1])/2)
if app_power.shape[1] < length:
    app_power = np.array([ left_padding_size*[1] + t[0] + right_padding_size*[1] for t in l ])
else:
    print(left_padding_size,right_padding_size)
    app_power = np.array([ t[0][left_padding_size*-1:right_padding_size] for t in l ])
    
```
